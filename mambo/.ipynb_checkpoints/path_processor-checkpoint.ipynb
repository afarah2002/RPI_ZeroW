```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
from pathlib import Path

font = cv2.FONT_HERSHEY_COMPLEX
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
```

```python
root = Path(".")

jpgfile = Image.open(root / "triangle.JPG")
pix = np.array(jpgfile)
plt.imshow(pix, cmap='gray')
```

```python
hsv = cv2.cvtColor(pix, cv2.COLOR_BGR2HSV)
lower_red = np.array([110,50,50]) 
upper_red = np.array([130,255,255])

mask_red = cv2.inRange(hsv, lower_red, upper_red)
# cv2.imshow('mask',mask)
plt.imshow(mask_red, cmap='gray',interpolation='bicubic')
plt.show()
```

```python
# _, threshold = cv2.threshold(mask,250,255,cv2.THRESH_BINARY_INV) 
_, contours, _ = cv2.findContours(mask_red, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
for cnt in contours:
    approx = cv2.approxPolyDP(cnt, 0.01*cv2.arcLength(cnt, True),True)
    cv2.drawContours(mask_red, [approx], 0, (0), 5)
    x = approx.ravel()[0]
    y = approx.ravel()[1]

    if len(approx) == 4:
        cv2.putText(mask_red, "Square", (1000,1000), font, 10, 255, 10, cv2.LINE_AA)
    
```

```python
plt.imshow(mask_red, cmap='gray',interpolation='bicubic')
plt.show()
```
